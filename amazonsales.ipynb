```python
DaySalesInsights = df.copy()
DaySalesInsights['Order Date'] = pd.to_datetime(df2['Order Date']).dt.date
```

```python
top10sales = DaySalesInsights.groupby('Order Date').sum().sort_values('Total Cost', ascending = False)
top10sales = top10sales.reset_index().head(10)
```

```python
sns.catplot(y = 'Total Cost', x = 'Order Date', data = top10sales, aspect = 2,palette='turbo',kind="bar")
plt.title('Top 10 Days When Sales Were Highest')
top10sales[['Total Cost']]
plt.show()
```

```python
plt.figure(figsize=(18,7))
plt.pie('Total Cost',labels='Order Date',data = top10sales,
        autopct='%1.2f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.title('Contribution Of Sales Amount Among 10 Days')
plt.legend(round(top10sales['Total Cost'],2), loc=7, fontsize = 'x-large')
plt.show()
```

```python
top10profits = DaySalesInsights.groupby('Order Date').sum().sort_values('Total Profit', ascending = False)
top10profits = top10profits.reset_index().head(10)
```

```python
sns.catplot(y = 'Total Profit', x = 'Order Date', data = top10profits, aspect = 2,palette='turbo',kind="bar")
plt.title('Top 10 Days When Profits Were Highest')
top10profits[['Total Profit']]
```

```python
plt.figure(figsize=(18,7))
plt.pie('Total Profit',labels='Order Date',data = top10profits,
        autopct='%1.2f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.title('Contribution Of Profits Among Top 10 Days')
plt.legend(round(top10profits['Total Profit'],2), loc=7, fontsize = 'x-large')
plt.show()
```
